```python
import scipy.io
import os
import numpy as np
import re
import pandas as pd

data_dir1 = os.listdir('data')
mapping_class = {'AD': 0,'MCI': 1,'SCI': 2}

df=dict()

columns=[i for i in range(1,901)]

for fdata_dir in data_dir1:
    data_list=[]
    df[fdata_dir] = pd.DataFrame(columns=columns)
    # dossiers alpha, beta, delta, theta
    data_dir2 = os.listdir('data/'+fdata_dir)
    for sdata_dir in data_dir2:
        # dossiers ad mci sci 
        dir_path = 'data/'+fdata_dir+'/'+sdata_dir
        for data_file in os.listdir(dir_path):
            if(re.search('\.mat',data_file)!=None):
                data_matrix_path = dir_path + '/' + data_file
                # .mat files
                data = scipy.io.loadmat(data_matrix_path)[fdata_dir]
                data = np.reshape(data,(900,))
                data = np.append(data,[mapping_class[sdata_dir]],axis=0)
                data_list.append(data)
    
    df[fdata_dir]=df[fdata_dir].append(data_list, ignore_index=True)
    df[fdata_dir] = df[fdata_dir].rename({900:'classe'},axis=1)
    df[fdata_dir].to_csv("data_csv/"+fdata_dir+".csv",sep=';')
```

```python
df_alpha = pd.read_csv('data_csv/ALPHA.csv', sep = ';').drop(['Unnamed: 0'], axis = 1)
df_beta = pd.read_csv('data_csv/BETA.csv', sep = ';').drop(['Unnamed: 0'], axis = 1)
df_delta = pd.read_csv('data_csv/DELTA.csv', sep = ';').drop(['Unnamed: 0'], axis = 1)
df_theta = pd.read_csv('data_csv/THETA.csv', sep = ';').drop(['Unnamed: 0'], axis = 1)
```
